```
def run_CentroidFold(input_file):
  import subprocess
  import re
  foldout, folderr = subprocess.Popen(['centroid_fold',input_file], stdin=subprocess.PIPE, stderr=subprocess.PIPE, stdout=subprocess.PIPE, universal_newlines=True).communicate()
  if folderr:
      raise 'Error'
  #print(foldout)
  # parse the results
  output_lines = foldout.strip().splitlines()
  rna_seq = output_lines[1]
  rna_structure = re.split('\s+', output_lines[2].strip())[0]
  #print("RNA_sequence : ",rna_sequence)
  #print("RNA_structure: ",rna_structure)
  return rna_seq,rna_structure
```

```
#@title Run prediction on bpRNA_TR0.fasta

import time
lines=open('bpRNA_TR0.fasta').readlines()
with open('CentroidFold_bpRNA_TR0.prediction',"w") as newFasta:
  for i in range(0,len(lines),2):
    id = lines[i].strip()
    seq = lines[i+1].strip()
    if i % 100 == 0:
        print(i,' finish!')
    !echo "$id" > 'test.fa'
    !echo "$seq" >> 'test.fa'
    try:
      (rna_sequence,rna_structure) = run_CentroidFold('test.fa')
    except:
      print("Failed for ", lines[i])
    header = id + "\t" + str(rna_sequence) + "\t" + str(rna_structure) + "\n"
    newFasta.write(header)
newFasta.close()
```
